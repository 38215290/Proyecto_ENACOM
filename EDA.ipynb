DATASET DE RELACION DE PROVINCIA CON BAF(BANDA ANCHA FIJA )

Esta métrica es importante para tener una visión más detallada del acceso a Internet fijo en cada provincia y para observar la tendencia de adopción de tecnologías de banda ancha en comparación con la banda angosta(Dial up) a lo largo del tiempo. También permite identificar áreas con alta demanda de banda ancha y áreas que podrían necesitar mejoras en la infraestructura para aumentar el acceso a Internet de alta velocidad.

En resumen, la métrica es simplemente el recuento de accesos a banda ancha y banda angosta para cada provincia y trimestre. Esta información es valiosa para analizar y comparar la disponibilidad y el uso de diferentes tipos de conexiones de Internet fijo en diferentes regiones

```python
df2=pd.read_csv("Dataset\Internet_BAF-por-provincia.csv")
pd.DataFrame(df2)
```

```python
#tamaño (cantidad de registros) del  dataset
len(df2)
```

```python
#se muestra el tipo de variable, los valores nulos y no nulos de cada columna
df2.info()
```

```python
#una transformacion fue que se reemplaza de la columna Banda ancha '.' por ' ' para luego convertir esos valores a numericos flotantes
df2['Banda ancha fija']=df2['Banda ancha fija'].str.replace('.', '').astype(float)
df2['Total']=df2['Total'].str.replace('.', '').astype(float)

```

```python
df_numerico=df2[['Banda ancha fija','Dial up','Total']]
```

```python
df_numerico.replace('- 0', np.nan, inplace=True)
```

```python
correlacion=df_numerico.corr()

#sns.heatmap(correlacion)
sns.heatmap(correlacion, cmap='coolwarm', annot=True, fmt=".2f", linewidths=1.0 ) 
plt.title("Correlation Heatmap") 
plt.show() 
```

```python
#se agrupa por por provincia y trimestre luego agrupados se saca un promedio de los valores agrupados de la columna Banda ancha fija
df_provincia = df2.groupby(['Provincia', 'Trimestre'])['Banda ancha fija'].mean().reset_index()

# Crear el gráfico de barras
fig, ax = plt.subplots(figsize=(12, 6))
for provincia in df2['Provincia'].unique():
    data = df_provincia[df_provincia['Provincia'] == provincia]
    ax.bar(data['Trimestre'], data['Banda ancha fija'], label=provincia)

ax.set_xlabel('Trimestre')
ax.set_ylabel('Banda ancha fija (promedio)')
ax.legend(title='Provincia', loc='upper left', bbox_to_anchor=(1, 1))
plt.title('Promedio de Banda ancha fija por Trimestre para cada Provincia')
plt.xticks(df2['Trimestre'].unique())
plt.show()
```

```python
#se agrupa por por provincia y año luego agrupados se saca un promedio de los valores agrupados de la columna Banda ancha fija
df_provincia_anio = df2.groupby(['Provincia', 'Año'])['Banda ancha fija'].mean().reset_index()

# Crear el gráfico de líneas para cada provincia 
fig, ax = plt.subplots(figsize=(14, 7))
for provincia in df2['Provincia'].unique():
    data = df_provincia_anio[df_provincia_anio['Provincia'] == provincia]
    ax.plot(data['Año'], data['Banda ancha fija'], label=provincia)

ax.set_xlabel('Año')
ax.set_ylabel('Banda ancha fija (promedio)')
ax.legend(title='Provincia', loc='upper left', bbox_to_anchor=(1, 1))
plt.title('Promedio de Banda ancha fija por Año para cada Provincia')
plt.xticks(rotation=45)
plt.show()
```

DATASET DE EL PROMEDIO DE LA CANTIDAD DE BAJADAS(Mbps) EN EL TRANSCURSO DE LOS AÑOS

La metrica que se evalua es la velocidad de descarga 

La rapidez con la que puede extraer datos de un servidor en Internet a su dispositivo. La mayoría de las conexiones están diseñadas para descargar mucho más rápido de lo que cargan. Esto se debe a que la mayor parte de la actividad en línea, como cargar páginas web o transmitir videos, consiste en descargas. La velocidad de descarga se mide en megabits por segundo ( Mbps ).



Mbps :
Megabits por segundo. Un megabit es 1 millón de bits de información. Esta es una medida estándar de la velocidad de Internet y 1 Mbps es 1000 veces más rápido que 1 Kbps. No debe confundirse con megabytes (MB), que es una medida de tamaño en lugar de ancho de banda. Debido a que un megabyte es más grande que un megabit, y para evitar confusiones, MB/s no se usa para indicar las velocidades de Internet.

DATASET DE ACCESO A INTERNET POR CADA 100 HOGARES

La métrica "Penetración de Internet fijo (accesos por cada 100 hogares)" es una medida utilizada para evaluar la adopción o la disponibilidad de servicios de Internet fijo en una determinada área o provincia. Esta métrica expresa la proporción de accesos al servicio de Internet fijo en relación con el número de hogares en esa área, expresado como un porcentaje.

La fórmula para calcular la penetración de Internet fijo por cada 100 hogares es:

Penetración de Internet fijo = (Número de accesos al servicio de Internet fijo / 100 hogares) * 100

```python
#visualizamos el dataset que posee en cada columna el promedio de bajada (Mbps) por año(trimestre)
df5=pd.read_csv("Dataset\Internet_Penetracion-por-100-hog-provincia.csv")
pd.DataFrame(df5)
```

definicion de hogar:Grupo de personas (emparentadas o no emparentadas entre sí) que comparten la misma vivienda, las que se asocian para proveer en común a sus necesidades alimenticias o de otra índole vital. Ahora en el dataset se puede observar que 

```python
#tamaño (cantidad de registros) del  dataset
len(df5)
```

```python
#se muestra el tipo de variable, los valores nulos y no nulos de cada columna
df5.info()
```

```python
trimestre=df5['Trimestre'].tolist()
```

```python
#se crea una nueva columna  periodo con el trimestre correspondiente y el año para mayor prolijidad
for indice,valor  in enumerate(trimestre):
    if valor == 1:
        df5.loc[indice, 'Periodo'] = 'Ene-Mar ' + str(df5.loc[indice, 'Año'])
    elif valor == 2:
        df5.loc[indice, 'Periodo'] = 'Abr-Jun ' + str(df5.loc[indice, 'Año'])
    elif valor == 3:
        df5.loc[indice, 'Periodo'] = 'Jul-Sep ' + str(df5.loc[indice, 'Año'])
    elif valor == 4:
        df5.loc[indice, 'Periodo'] = 'Oct-Dic ' + str(df5.loc[indice, 'Año'])
```

```python
df5
```

```python
#una transformacion fue que se reemplaza de la columna Accesos por cada 100 hogares ',' por ' . ' para luego convertir esos valores a numericos flotantes
df5['Accesos por cada 100 hogares'] = df5['Accesos por cada 100 hogares'].str.replace(',', '.')
```

```python
#se cambia el tipo de variable a la columna Accesos por cada 100 hogares
df5['Accesos por cada 100 hogares']=df5['Accesos por cada 100 hogares'].astype(float)
```

```python
#valores mayores al 100 % es porque en capital federal el numero de accesos a intenet excede el numero  de hogares ya que se puede tener 
# varias conexiones en un mismo hogar(edificio) recordando la siguiente definicion Hogar:grupo de personas (emparentadas o no emparentadas entre sí) que comparten la misma vivienda,
# las que se asocian para proveer en común a sus necesidades alimenticias o de otra índole vital
atipicos =df5['Accesos por cada 100 hogares']>100
df5[atipicos]
```

```python
#se agrupa por por provincia y año luego agrupados se saca un promedio de los valores agrupados de la columna Banda ancha fija
df_provincia_anio = df5.groupby(['Provincia', 'Año'])['Accesos por cada 100 hogares'].mean().reset_index()

# Crear el gráfico de líneas para cada provincia 
fig, ax = plt.subplots(figsize=(14, 7))
for provincia in df5['Provincia'].unique():
    data = df_provincia_anio[df_provincia_anio['Provincia'] == provincia]
    ax.plot(data['Año'], data['Accesos por cada 100 hogares'], label=provincia,marker='o', linestyle='-')

ax.set_xlabel('Año')
ax.set_ylabel('por cada 100 hogares')
ax.legend(title='Provincia', loc='upper left', bbox_to_anchor=(1, 1))
plt.title('Promedio de acceso a internet por cada 100 hogares')
plt.xticks(rotation=45)
plt.show()
```
